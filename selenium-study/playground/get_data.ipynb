```python
import tabula as tb
import pandas as pd
```

```python
# Set the file path
word_list_file_path = '../data/nihongo_no_mori.pdf'

# Load the file at a specific page
raw_word_list_df = tb.read_pdf(word_list_file_path, pages='12-21', stream=True, pandas_options={'header': None})

print(len(raw_word_list_df[0]))
raw_word_list_df[0].head(15)
```

```python
raw_word_list_df
```

```python
# Drop unnecessary columns
clean_word_list_df = raw_word_list_df[0].drop(columns=[1,3,4])

# Drop unnecessary rows
clean_word_list_df.drop(range(2), axis=0, inplace=True)
clean_word_list_df = clean_word_list_df.reset_index(drop=True)

# Reset column name
clean_word_list_df.columns = range(clean_word_list_df.shape[1])

clean_word_list_df.head(len(clean_word_list_df))
```

```python
# Removing the squares from the kanji rows [0, 2] and merging
# Also creates a new series for the corresponding meaning
kanji_row = pd.concat([clean_word_list_df[0].str[2:], clean_word_list_df[2].str[2:]], ignore_index=True)
kanji_meaning = pd.concat([clean_word_list_df[1], clean_word_list_df[3]], ignore_index=True)

# Create the final df
# final_word_list = pd.DataFrame(data=[kanji_row, kanji_meaning], columns=['word', 'reading'])
final_word_list = pd.DataFrame({
  'word': kanji_row
  , 'reading': kanji_meaning
})

print(len(final_word_list))
final_word_list.head()
```

```python
# final_word_list.head(len(final_word_list))

import time

# Save final df as an excel file
final_word_list.to_excel(
  '../data/res/testResult1.xlsx'
  , index=False
  , sheet_name='result' + '_' + time.strftime('%y%m%d_%H%M%S', time.localtime())
)
```
